# Has-a 포함관계
- 하나의 클래스가 다른 클래스를 속성으로 가지고 있는 경우
- A has a B 성립하는 경우

```python
class Engine:
    def __init__(self, horse_power: int):
        self.horse_power = horse_power

    def start_engine(self):
        print(f"{self.__horse_power}마력 엔진의 시동을 겁니다.")


    def stop_engine(self):
        print("엔진 시동을 끕니다.")


class Car:
    def __intit__(self, name:str, engine: Engine):
        self.name = name

        # Car 클래스가 Engine을 속성으로 가지고 있다.
        # == Car has an Engine
        self.__engine = engine
```

# Is-a 상속관계
* 하나의 클래스가 다른 클래스를 상속하는 경우
* A is a B 성립하는 경우

## Has-a와 Is-a 선택 기준

상속은 `A is a B`가 자연스럽게 성립할 때 사용한다.
포함관계는 `A has a B`가 자연스럽게 성립할 때 사용한다.

예를 들어 `Taxi is a Car`는 자연스럽지만, `Car is an Engine`은 어색하다.
따라서 자동차가 엔진 기능을 재사용해야 한다면 상속보다 포함관계가 더 적절하다.


```python
class Taxi(Car):
    def __init__(self, name: str, engine: Engine, taxi_type: str):
        super().__init__(name, engine)
        self.taxi_type = taxi_type  # 개인/회사 택시 여부


taxi = Taxi('소나타', Engine(200), '개인')


print('isinstance(taxi, Taxi):', isinstance(taxi, Taxi))
print('isinstance(taxi, Car):', isinstance(taxi, Car))


class Bus(Car):
    pass


class Truck(Car):
    pass
```
