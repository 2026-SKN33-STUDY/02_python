# for 반복문

- `for` 반복문은 iterable 객체의 요소를 하나씩 꺼내 실행한다.
- `str`, `list`, `tuple`, `dict`, `set`, `range` 등이 대표적인 iterable 객체이다.
- 반복 횟수가 정해져 있거나 컬렉션의 모든 요소를 순회할 때 자주 사용한다.

## 1. 문자열, 리스트, 튜플 순회

문자열은 문자 단위로, 리스트와 튜플은 요소 단위로 순회한다.

```python
text: str = 'hello world'
for ch in text:
     print(ch)

lst: list[int] = [10,20,30]
for num in lst:
    print(num)

tpl: tuple[int] = (30, 35, 40)
for num in tpl:
    print(num)
```

## 2. dict와 set 순회

`dict`를 그대로 순회하면 key가 나온다. `items()`를 사용하면 key와 value를 함께 꺼낼 수 있다.

`set`은 순서가 없는 자료형이므로 출력 순서는 실행 환경에 따라 달라질 수 있다.

```python
from typing import Any

# type hint - dict[str, Any]에서 Any
# - 어떤 타입이든 작성 가능하다는 것을 나타내는 표현(클래스)
user_info: dict[str, Any] = {
    'name': '홍길동',
    'age': 23,
    'hobby': ['독서', '유튜브']
}

# dict 순회 -> key 순회, value 순회, item 순회(+ unpacking)
# for A in B
# A: 객체에서 꺼내와 저장할 변수
# B: 순회가능한 객체


for key in user_info:
    print(f"{key}: {user_info[key]}")

for key, value in user_info.items():
    print(f"{key}: {value}")

print('----- set 순회: if를 곁들인 ------')
nums: set[int] = {1, 2, 3,3, 4,5, 6,7 ,3,3232,555, 2,1, 9, 7,5}

# 결과 저장용 list
results: list[int] = []

for num in nums:
    if num % 2 == 0:
        results.append(num)

print(results)
```

## 3. range 객체

`range(start, end, step)`은 정수 범위를 표현하는 객체이다.

- `start`: 시작값
- `end`: 끝값, 포함되지 않음
- `step`: 증가 또는 감소 간격

```python
rng = range(1,5,1)
print(rng, type(rng))

for num in rng:
    print(num)

print('---- range 객체를 이용한 정수 list 만들기 ----')
rng2 = range(1, 11)
nums: list[int] = list(range(1, 11, 1))
print("nums: ", nums)

print('--- 1부터 100까지 모든 정수 합 구하기 ---')
n_sum = 0
for n in range(1, 101, 1):
    n_sum += n
print('1부터 100까지의 합: ', n_sum)

print('----- 1=1부터 100까지 홀수의 합 구하기 -----')
n_sum = 0
for n in range(1, 101, 1):
    if n % 2 == 1:
        n_sum += n


print("홀수 합: ", n_sum)
```

```python
print('---- 2단 출력하기 ----')
dan: int = 2
for num in range(1, 10):
    print(f'{dan} x {num} = {dan * num}')
```

```python
print('---- 구구단 출력하기(2단~9단) ----')
for dan in range(2, 10):
    for num in range(1, 10):
        print(f'{dan} x {num} = {dan * num}')
    print('-'*20)


```

## 4. 중첩 반복문

반복문 안에 반복문을 넣으면 행과 열처럼 2차원 구조를 만들 수 있다. 구구단과 별찍기 같은 예제에 자주 사용한다.

```python
print('---- 중첩 반복문 예시 ----')
for a in range(1, 4):
    for b in range(1, 3):
        print(a, b)
```

```python
print('--- range를 이용한 반복 횟수 설정 ---')
print('🍔 3행 5열 출력')

for row in range(3):
    for col in range(5):
        print("🍔", end='')
    print() # 줄바꿈
```

## 5. zip과 enumerate

- `zip()`은 여러 iterable을 같은 순서끼리 묶는다.
- `enumerate()`는 인덱스와 요소를 함께 꺼낸다.

```python
print('----- zip 객체 -----')
names: list[str] = ['a', 'b', 'c']
nums : list[int] = [10, 20, 30]

z = zip(names, nums)
print(z, type(z))

for name, num in z:
    print(name, num)


print('----- enumerate 객체 -----')
foods: list[str] = ['제육', '마녀스프', '마라탕', '김볶밥']

for index, food in enumerate(foods):
    print(f'{index}번 인덱스: {food}')



```

## 6. 반복문으로 평균 구하기

반복문을 사용하면 리스트나 딕셔너리 리스트에 들어 있는 값을 누적해서 평균을 계산할 수 있다.

```python
print('---- list 에 담긴 값 평균 구하기 ----')
scores: list[int] = [55, 78, 99, 100, 60]

score_sum = 0

for score in scores:
    score_sum += score
# print(score_sum)

score_avg = score_sum / len(scores)
print("평균: ", score_avg)
print("평균: ", sum(scores)/len(scores))
# 숫자 list, set, tuple일 때만 가능
```

```python
print('----- dict에서 점수 평균 구하기 -----')

students: list[dict[str, object]] = [
    {'name': '홍길동', 'score': 100},
    {'name': '신사임당', 'score': 77},
    {'name': '이순신', 'score': 85},
]

print(students)


# list 순회
# students[0].get('score')

sum: int = 0
for std in students:
    sum += std.get('score')

avg = sum / len(students)
print('평균: ', round(avg, 2))
# round(소수점): 반올림

for index in range(len(students)):
    sum += students[index].get('score')

print(sum)


```
